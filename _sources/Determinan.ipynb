1. Determinan dari matriks identitas adalah 1
2. Misal Determinan AB merupakan matriks bujur sangkar maka determinan matriks A dapat dikalikan dengan determinan matriks B

determinan adalah kumpulan skalar dari semua matrix

```python
import numpy as np
a= np.array([[2,3],[3,4]])
print ('matrix 2x2')
print (a)
b=np.array([[4,5,6],[2,3,4],[3,4,2]])
print ('matrix 3x3')
print(b)
c=np.array([[1,0],[0,1]])
print('matrix identitas 2x2')
print (c)
d=np.array([[1,0,0],[0,1,0],[0,0,1]])
print ('matrix identitas 3x3')
print (d)

```

1. sifat sifat determinan

```python
Ic = np.linalg.det(c)
Id = np.linalg.det(d)
print (Ic)
print (Id)
```

```python
inversA = np.linalg.inv(a)
print ('invers A')
print (inversA)
```

```python
inversB = np.linalg.inv(b)
print ('invers B')
print (inversB)
```

2. determinan dari matrix A B = Matrix identitas

```python
a1 =  np.linalg.det(a)
DetA= np.linalg.det(inversA)

kali =a1 * DetA
print (kali)
```

3. Determinan dari (a invers) =1/a

```python
import numpy as np
e = np.array ([[1,2,3],[2,3,4],[2,3,5]])
print ('matrix e 3x3')
print (e)
f = np.array ([[1,2,1],[2,1,2],[1,1,2]])
print ('matrix f 3x3')
print (f)
g =e@f
detg= np.linalg.det(g)
print ('matrix g')
print (g)
print ('determinan matrix g')
print (detg)

detEF =np.linalg.det(e)*np.linalg.det(f)
print ('determinan matrix f')
print (f)
```
