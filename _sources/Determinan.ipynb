# Determinan Matriks

Determinan merupakan

Sifat sifat determinan matriks :

1. Determinan dari matriks identitas adalah 1
2. Misal Determinan AB merupakan matriks bujur sangkar maka determinan matriks A dapat dikalikan dengan determinan matriks B
3. MEnukar baris dari matriks bisa merubah tanda dari determinan
4. jika matriks a dikalikan dengan skalar ,maka determinan dari ka bisa di cari dengan cara skalar dikalikan dengan orde dari matriks a,lalu dikalikan dengan determinan a.
5. melakukan operasi baris elementer tidak merupah nilai determinan dari matriks tersebut
6. Determinan matriks a tranpose sama dengan determinan matriks a.
7. Jika skalar dikalikan dengan baris 1 dari matriks pertama maka ,determinan ka bisa dicari dengan skalar dikali determinan martriks.
8. determinan matriks segitiga dicari dengan mengalikan setiap elemen yang ada di matriks diagonalnya saja

determinan adalah kumpulan skalar dari semua matrix

```python
import numpy as np
a= np.array([[2,3],[3,4]])
print ('matrix 2x2')
print (a)
b=np.array([[4,5,6],[2,3,4],[3,4,2]])
print ('matrix 3x3')
print(b)
c=np.array([[1,0],[0,1]])
print('matrix identitas 2x2')
print (c)
d=np.array([[1,0,0],[0,1,0],[0,0,1]])
print ('matrix identitas 3x3')
print (d)

```

1. sifat sifat determinan

```python
Ic = np.linalg.det(c)
Id = np.linalg.det(d)
print (Ic)
print (Id)
```

```python
inversA = np.linalg.inv(a)
print ('invers A')
print (inversA)
```

```python
inversB = np.linalg.inv(b)
print ('invers B')
print (inversB)
```

```python
a1 =  np.linalg.det(a)
DetA= np.linalg.det(inversA)

kali =a1 * DetA
print (kali)
```

## tugas mencari determinan matriks 4x4 dengan cara ekspansi kofaktor
1. contoh matrix 4x4 :

$\begin{matrix}
A = \begin{bmatrix}
    4 & 5 & 6 & 7\\
    2 & 2 & 2 & 2\\
    4 & 3 & 2 & 3\\
    3 & 3 & 6 & 3
  \end{bmatrix}
\end{matrix}
$

2. selanjutnya tulis rumus determinan disini, mengambil baris ke-2 :

$ det(A) = a_{21} c_{21} + a_{22} c_{22}  + a_{23} c_{23} + a_{24} c_{24} $

3.selanjutnya tentukan kofaktor nya disini dari matriks A


$c_{21} = (-1)^{2+1} det(M_{21})$  
$c_{22} = (-1)^{2+2} det(M_{22})$  
$c_{23} = (-1)^{2+3} det(M_{23})$  
$c_{24} = (-1)^{2+4} det(M_{24})$

4. setelah itu kita akan cari satu-persatu determinan minor tersebut dari matriks A

5. pertama : $M_{21}$

$\begin{matrix}
  det(M21) = \begin{bmatrix}
    5 & 6 & 7 \\
    3 & 2 & 3 \\
    3 & 6 & 3
  \end{bmatrix}
\end{matrix}
$  

selanjutnya tulis rumus determinan, dan mengambil baris ke-2  
$det(M21) = a_{21}c_{21} + a_{22}c_{22} + a_{23}c_{23}$   

selanjutnya tentukan kofaktor dari matriks yang $M_{21}, M_{22}, M_{23}$  
* $c_{21} = (-1)^{2 + 1} det(KM21)$

$
\begin{matrix}
  det(KM21) = \begin{bmatrix}
    4 & 5 \\
    5 & 4
  \end{bmatrix}
\end{matrix}
$  
$det(KM21) = 6.7 - 6.3 = 42 - 18 = -24$

hasil
$c_{21} = -1*-24$

* $c_{22} = (-1)^{2+2}det(KM22)$


$
\begin{matrix}
  det(KM22) = \begin{bmatrix}
    5 & 3 \\
    3 & 7
  \end{bmatrix}
\end{matrix}
$  
$det(KM22) = 5.7 - 3.3 = 35 - 9 = 26$  

hasil
$c_{22} = -1*26$  

* $c_{23} = (-1)^{2+3}det(KM23)$  

$
\begin{matrix}
  det(KM23) = \begin{bmatrix}
    5 & 6 \\
    3 & 6
  \end{bmatrix}
\end{matrix}
$  
$det(KM23) = 5.6 - 6.3 = 30 - 18 = 12$  


hasil
$c_{23} = -1*12$  


**hasil akhir dari $M_{21}$**  
$det(M21) = a_{21}c_{21} + a_{22}c_{22} + a_{23}c_{23}$  
$det(M21) = 2.(-1).(-9) + 2.1.2 + 2.(-1).7$  
$det(M21) = 18 + 4 - 14 = 8$










```python
import numpy as np
M21 = np.array([[3,4,5], [2,2,2], [2,5,4]])
detM21 = np.linalg.det(M21)
print(round(detM21))
```

$1.(-1).8 + 1.1.17 + 1.(-1).0 + 1.(-1).(-15) =$
$-8 + 17 + 0 + 15 = 4$
